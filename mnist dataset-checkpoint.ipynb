```python
#its shape is(60000*785) which my computer was unable to compute and giving memory error,
#so i have reduced the length of the data

df_data= data.iloc[0:20000,1:]
df_label=data.iloc[0:20000,0]
```

```python
data_train,data_test,label_train,label_test=train_test_split(df_data,df_label,test_size=0.2,random_state=4)
```

```python
#data_train.head(123)
```

```python
#label_train.head(123)
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(data_train,label_train)
```

```python
data_test.head()
```

```python
label_test.head()
```

```python
pred= rf.predict(data_test)
```

```python
pred

```

```python
s=label_test.values
s
```

```python
count=0
for i in range(len(label_test)):
    if s[i]==pred[i]:
        count+=1
print( "Total correct answer is {}".format(count))
```

```python
k=len(label_test)
print("total no. of prediction is {}".format(k))
```

```python
print('accuracy is {}%'.format((count*100)/k) )
```
